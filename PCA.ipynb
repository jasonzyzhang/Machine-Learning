```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### (a)

```python
problem4_x_train = np.load('problem4_x_train.npy')
problem4_y_train = np.load('problem4_y_train.npy')
```

```python
problem4_x_train.shape
```

```python
x1 = problem4_x_train[:,0]
x2 = problem4_x_train[:,1]
```

```python
plt.scatter(x1,x2,c=problem4_y_train,alpha=0.6)
plt.show()
```

```python
# the datapoints seem to be contained within the left-bottom corner of the plot, mostly within the range
# for x1 between 0.0 and 0.1 and the range for x2 between 0.0 and 0.1
# this is not a useful projection of the data
```

### (b)

```python
A = np.random.normal(0, 1, size=(2, 729))
```

```python
A
```

```python
x_ = A.dot(problem4_x_train.T)
```

```python
plt.scatter(x_[0],x_[1],c=problem4_y_train)
plt.show()
```

```python
# the class labels are pretty random and show no clear pattern based on the product of the
# random Gaussian matrix A with x1 and x2
```

### (c)

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(problem4_x_train)
```

```python
P = pca.components_
```

```python
x_PCA = P.dot(problem4_x_train.T)
```

```python
plt.scatter(x_PCA[0],x_PCA[1],c=problem4_y_train)
plt.show()
```

```python
# the data points seem to be arranged well based on their class labels
# this shows that PCA is helpful in accurately illustrating the class labels
# based on their features' principal components
```
